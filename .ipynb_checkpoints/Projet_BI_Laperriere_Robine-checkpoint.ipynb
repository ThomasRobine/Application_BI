# Introduction

L’objectif de ce projet est de mettre en place différentes approches issues de la fouille de données
afin de pouvoir prédire les démissions de sociétaires d’un organisme de prêt bancaire. Il s’agira
donc de pouvoir appliquer différents algorithmes de classification, avec objectif principal de
pouvoir expliquer les choix de la décision de ces algorithmes. Cela permettra de mettre en valeur
les caractéristiques conduisant à la démission, ou au contraire, au fait de rester dans l’organisme
bancaire. Dans un second temps, vous explorerez des approches non-supervisées de clustering
permettant de faire apparaitre des groupes d’utilisateurs ensemble.

Un certain nombre de variables ont été relevées sur les sociétaires démissionnaires d'un organisme
bancaire (i.e. clients ayant quitté la banque), ainsi que sur un échantillon de sociétaires actuels nondémissionnaires. Une méthode d'analyse prédictive permet de modéliser le fait de démissionner ou
pas à partir des valeurs des variables pour chaque sociétaire. Un score est produit, qui sera ici égal à
la probabilité d'être démissionnaire.

Bien évidemment, si le modèle est pertinent, cette probabilité sera forte pour les démissionnaires,
faible pour les sociétaires actuels. Mais il peut arriver que ce score soit élevé pour certains
sociétaires (non-démissionnaires). On peut alors en conclure que le profil de ce sociétaire est proche
de celui des démissionnaires, et donc que le risque de démission de ce sociétaire est élevé.

On a ainsi constitué un score d'attrition, ou risque de démission, calculable pour chaque sociétaire
de la banque (et actualisable en fonction de l'évolution de sa situation). Ce score peut servir de base
à des campagnes de gestion de la relation-client (prévention du départ des sociétaires).

### Demission data

```python
print(demission_data.shape)
print(demission_data.dtypes)
```

Nous pouvons voir que `demission_data` possède 19 attributs dont 8 qui sont de type `catégoriel` et 11 de type `numérique`.

```python
print(demission_data.isna().sum())
```

Nous pouvons remarquer que l'attribut `rangadh` est le seul attribut à posséder des valeurs non-renseignées, qui sont au nombre de 2584.

```python
print(demission_data.isin(['?']).sum(axis=0))
```

Nous pouvons voir qu'aucun attribut ne comporte de '?' parmis ses valeurs.

```python
print(demission_data.isin([0]).sum(axis=0))
```

Nous pouvons remarquer plusieurs choses.

Tout d'abord, en ce qui concerne l'attribut `MTREV`. Nous pouvons voir qu'un total de 27050 demissionaires sur 30332 ont 0 pour montant de revenue. Cela semble abérrant et nous pensons donc qu'il s'agit là de valeurs non renseignées.

Ensuite, l'attribut `NBENF`. Nous pouvons observer le fait que 27050 démissionnaires ne semblent pas avoir d'enfants.

Pour l'attribut `CDTMT`, rien de grave étant donné qu'il s'agit d'une variable catégorielle.

Enfin, pour l'attribut `rangadh`, pareil que précédemment, il s'agit de valeurs manquantes.

```python
print(demission_data['AGEAD'].where(demission_data['NBENF'] == 0).value_counts())
```

Nous pouvons qu'il va y avoir des valeurs atypiques. En effet, nous sommes à même de nous demander si les les personnes de 80+ ans n'ont vraiment pas d'enfants. Ou alors, si cela comprend les clients qui n'ont pas d'enfant eux aussi sociétaires de la banque.
Il faudra éclaircir ce point.

```python
names = demission_data.columns

for name in names:
    if name in ['NBENF', 'AGEAD', 'agedem']:
        print('Mean :\n', demission_data[name].mean())
        print('Median :\n', demission_data[name].median())
        print('Name:', name,', dtype:',demission_data[name].dtypes, '\n')
    elif name in ['CDSEXE', 'CDSITFAM', 'CDTMT', 'CDDEM', 'ANNEE_DEM', 'CDMOTDEM', 'CDCATCL']:
        print(demission_data[name].value_counts(), '\n')
    elif name in ['DTADH', 'DTDEM']:
        date_shortened = []
        for date in demission_data[name]:
            date_shortened.append(date.split('-')[:2])
        date_shortened = pd.DataFrame(date_shortened)
        date_shortened.columns=["Year", "Month"]
        print(date_shortened.value_counts())
        print('Name:', name, ', dtype:', demission_data[name].dtypes, '\n')
```

```python
import matplotlib
import matplotlib.pyplot as plt

features = ['NBENF', 'AGEAD', 'agedem', 'CDSEXE', 'CDSITFAM', 'CDTMT', 'CDDEM', 'CDMOTDEM', 'CDCATCL']
x = demission_data[features]
y = demission_data['ANNEE_DEM']

fig = plt.figure()
cmap = matplotlib.cm.get_cmap('gnuplot')
scatter = pd.plotting.scatter_matrix(
    x, 
    c = y, 
    marker='o', 
    s=40, 
    hist_kwds={'bins':15}, 
    figsize=(15,15), 
    cmap=cmap
)
plt.suptitle('Scatter-matrix for each input variable')
```

```python
# print histogram for each attribute with belonging to classes

features = ['NBENF', 'AGEAD', 'agedem', 'CDSEXE']
for attribut in features:
    fig = plt.figure()
    pd.DataFrame({k: v for k, v in demission_data.groupby('ANNEE_DEM')[attribut]}).plot.hist(stacked=True)
    plt.suptitle(attribut)
```
